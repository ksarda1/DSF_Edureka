```python
# importing the required libraries

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# reading the data

data = pd.read_csv('breast-cancer-data.csv')
data.isna().sum()
```

So we don't have any null values in the data. Let's start analysing the data

```python
data.dtypes
```

```python
data
```

The 'id' and 'diagnosis' columns are not necessary for our Dimension Reductionality purpose. So we work on rest of the features

```python
data1 = data.drop(['id','diagnosis'],axis=1)
data1.describe()
```

Since the features of data are not on proper scale, we sclae the data first

```python
scale = StandardScaler()
data2 = scale.fit_transform(data1)
```

Now let's do PCA by finding optimal number of components

```python
pca = PCA()
pca.fit(data2)
plt.plot(pca.explained_variance_ratio_.cumsum())
```

So we can see that around 90% of variance is explained by just 5 Principal Components

So let's find out the data in terms of those 5 components

```python
pca = PCA(n_components=5)
pca_data = pca.fit_transform(data2)
pca_data = pd.DataFrame(pca_data)
```

Adding the 'id' and 'diagnosis' features back to the new reduced data and saving the data to .csv format

```python
pca_data['id'] = data['id']
pca_data['diagnosis'] = data['diagnosis']
pca_data.to_csv('PCA_data.csv')
```
