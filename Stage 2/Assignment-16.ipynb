```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1.a 

```python
prisoner = pd.read_csv('prisoners.csv')
print('First five rows',end='\n')
print(prisoner.head(),end='\n\n\n')
print('Last five rows',end='\n')
print(prisoner.tail())
```

# 1.b 

```python
prisoner.describe
```

```python
prisoner.loc[prisoner.isin([0]).any(axis=1)]
```

# 2.a

```python
prisoner_tot = prisoner.copy()
prisoner_tot['Total_Benefitted'] = prisoner_tot[['No. of Inmates benefitted by Elementary Education',
                                        'No. of Inmates benefitted by Adult Education',
                                        'No. of Inmates benefitted by Higher Education',
                                        'No. of Inmates benefitted by Computer Course']].sum(axis=1)
prisoner_tot
```

# 2.b 

```python
prisoner.loc['totals'] = prisoner[['No. of Inmates benefitted by Elementary Education',
                               'No. of Inmates benefitted by Adult Education',
                                'No. of Inmates benefitted by Higher Education',
                                'No. of Inmates benefitted by Computer Course']].sum()
prisoner
```

# 3.a 

```python
plt.figure(figsize=(15,5))
plt.bar(prisoner_tot['STATE/UT'],prisoner_tot['Total_Benefitted'])
plt.title('Statewise total Inmates',fontsize=25)
plt.xlabel('State',fontsize=10)
plt.ylabel('Total Inmates',fontsize=10)
plt.yticks(fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.show()
```

```python
print('Andhra Pradesh has the maximum number of beneficiaries')
```

# 3.b 

```python
x = prisoner[['No. of Inmates benefitted by Elementary Education',
                               'No. of Inmates benefitted by Adult Education',
                                'No. of Inmates benefitted by Higher Education',
                                'No. of Inmates benefitted by Computer Course']]
plt.figure(figsize=(10,5))
plt.pie(x.loc['totals'],labels=['No. of Inmates benefitted by Elementary Education',
                               'No. of Inmates benefitted by Adult Education',
                                'No. of Inmates benefitted by Higher Education',
                                'No. of Inmates benefitted by Computer Course'],autopct='%1.1f%%')
plt.title('Ratio of various Inmate types',fontsize=20)
plt.show()
```
