```python
from PIL import Image
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
im = Image.open("dogs.jpeg")
im
```

```python
size = im.size
size
```

```python
raw_data = list(im.getdata())
data = pd.DataFrame(raw_data)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
clus_data = kmeans.cluster_centers_[kmeans.labels_]
clus_data /= 255
clus_data = clus_data.reshape(size[1],size[0],3)
```

```python
plt.imshow(clus_data)
```

```python
dom_colors = kmeans.cluster_centers_/255

plt.imshow(dom_colors[0].reshape(1,1,3))
plt.show()
plt.imshow(dom_colors[1].reshape(1,1,3))
plt.show()
plt.imshow(dom_colors[2].reshape(1,1,3))
plt.show()
```
