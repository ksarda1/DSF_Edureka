```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import mean_squared_error
```

```python
raw_data = pd.read_csv('zoo.csv')
raw_data
```

```python
raw_data.class_type.unique()
```

```python
len(raw_data.class_type.unique())
```

```python
data = raw_data.drop(['animal_name','class_type'],axis=1)
```

```python
agg = AgglomerativeClustering(n_clusters=7)
preds = agg.fit_predict(data)
```

```python
mean_squared_error(raw_data.class_type,preds)
```
