```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 1. 

```python
hurricanes = pd.read_csv('Hurricanes.csv')
plt.bar(hurricanes.Year,hurricanes.Hurricanes)
plt.title('Bar plot of number of Hurricanes per year',fontsize=20)
plt.xlabel('Year',fontsize=15)
plt.ylabel('number of Hurricaes',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

# 2. 

```python
temp = pd.read_csv('CityTemps.csv')
plt.hist(temp.Moscow)
plt.title('Histogram of temperature for Moscow',fontsize=20)
plt.xlabel('Temperature',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
plt.hist(temp['San Francisco'])
plt.title('Histogram of temperature for San Francisco',fontsize=20)
plt.xlabel('Temperature',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

# 3. 

```python
data = pd.read_csv('sample-salesv2.csv')
```

```python
data.describe()
```

```python
data['unit price'].describe()
```

```python
new_dataframe = data[['name','net_price','date']]
new_data = new_dataframe.groupby('name')
```

```python
plt.plot(new_data.sum()/1000)
plt.title('Sales by Customer',fontsize=20)
plt.xlabel('Customer',fontsize=15)
plt.ylabel('Sales amount (in thousand dollars)',fontsize=15)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
        'female': [0, 1, 1, 0, 1],
        'age': [42, 52, 36, 24, 73], 
        'preTestScore': [4, 24, 31, 2, 3], 
        'postTestScore': [25, 94, 57, 62, 70]}
data = pd.DataFrame(data)
plt.scatter(data.preTestScore,data.postTestScore,s=data.age)
plt.show()
```
