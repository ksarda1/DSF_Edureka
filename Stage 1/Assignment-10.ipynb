```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('BigMartSalesData.csv')
```

# 1. Plot of Total Sales Per Month for Year 2011

```python
data_2011 = data.loc[data.Year == 2011]
sales_2011 = data_2011.groupby('Month').sum()/1000000
month = ['January','February','March','April','May','June','July','August','September','October','Novenmber','December']

plt.plot(month,sales_2011.Amount)
plt.title('Sales in 2011',fontsize=20)
plt.xlabel('Month',fontsize=15)
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.ylabel('Sales (in million dollars)',fontsize=15)
plt.show()
```

#### Minimum Sale is for February

# 2. Bar chart of Sales of year 2011

```python
plt.bar(month,sales_2011.Amount)
plt.title('Sales in 2011',fontsize=20)
plt.xlabel('Month',fontsize=15)
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.ylabel('Sales (in million dollars)',fontsize=15)
plt.show()
```

#### Clearly Bar Chart is better

# 3. Pie chart for country wise Sales

```python
country_sales_2011 = data_2011.groupby('Country').sum()
plt.pie(country_sales_2011.Amount,labels=country_sales_2011.index)
plt.title('Pie chart for country wise Sales',fontsize=20)
plt.show()
```

#### United Kingdom has the highest sales

# 4. Scatter Plot for the invoice amounts

```python
plt.scatter(data_2011.index,data_2011.Amount,c='b')
plt.ylim(0,5000)
plt.ylabel('Invoice amount',fontsize=20)
plt.title('Plot for the invoice amounts',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

#### Most of the invoices are below the amount 2000
