```python
# importing the required libraries

import nltk
from nltk.corpus import shakespeare
import string
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer 
from nltk.tokenize import word_tokenize
```

```python
# reading all the shakspeare documents as shaks1, shaks2, shaks3

shaks1 = nltk.corpus.gutenberg.words('shakespeare-hamlet.txt')
shaks2 = nltk.corpus.gutenberg.words('shakespeare-caesar.txt')
shaks3 = nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')
```

```python
# function to build paragraph from given list of words

def para(doc) :
    t = nltk.Text(doc)
    stop_words = stopwords.words('english')
    punct =string.punctuation

    data = t.tokens
    mean_words =[]
    for word in data:
        if word not in punct:
            if word.lower() not in stop_words:
                mean_words.append(word)
    return mean_words
```

```python
# function to find the most frequent bigrams in given document
# function takes an argument 'freq' which signifies top frequent bigrams to find

def freq_bi(doc,freq) :
    t = nltk.Text(doc)
    string = ''
    for i in t.tokens :
        string += i
        string += ' '
    k = list()
    k.append(string)
    vect2 = CountVectorizer(ngram_range=(2,2),max_features=freq)
    vect2.fit_transform(k)
    return vect2.vocabulary_.keys()
```

```python
# list of meaningful words from shaks1 document
para(shaks1)
```

```python
# list of meaningful words from shaks2 document
para(shaks2)
```

```python
# list of meaningful words from shaks3 document
para(shaks3)
```

```python
# top 10 frequent bigrams from original document shaks1
freq_bi(shaks1,10)
```

```python
# top 10 frequent bigrams from original document shaks2
freq_bi(shaks2,10)
```

```python
# top 10 frequent bigrams from original document shaks3
freq_bi(shaks3,10)
```

```python
# top 10 frequent bigrams from meaningful words of document shaks1
freq_bi(para(shaks1),10)
```
