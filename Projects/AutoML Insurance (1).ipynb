```python
data = pd.read_csv('train(1).csv')
data = data.drop(['id'],axis=1)
```

Data has 59 columns out of which 'id' column is neglected for model building. 'Target' variable is the prediction variable. There are Continous, Categorical and Binary Variables in the data. All the Data Preprocessing is done. 

Target Variable with Target=1 is just 3.65% in the data. Hence Data is imbalanced

```python
cat_var = 0
bin_var = 0
for i in data.columns :
    if data[i].dtypes != float :
        if len(data[i].unique()) == 2 :
            bin_var += 1
        else :
            cat_var += 1

print('Number of Categorical variable are ',cat_var)
print('Number of Binary Variable are ',bin_var)
```

```python
float_var = []
cat_var = []
bin_var = []
for i in data.columns :
    if data[i].dtypes != 'float' :
        if len(data[i].unique()) == 2 :
            bin_var.append(i)
        else :
            cat_var.append(i)
    else :
        float_var.append(i)
```

If we Undersample, Dataset size should be around 180000, which leads to lose of a lot of information.
So best way is to Oversample.

```python
from sklearn.utils import resample
```

```python
Xr = resample(data.loc[data.target==1],n_samples=80000)
data0 = data.loc[data.target==0]
data_new = data0.append(Xr)
data_new.describe()
```

```python
miss_data = []
for i in data.columns :
    x = data[i]==-1
    if x.sum() :
        miss_data.append([x.sum(),i])

miss_data.sort(reverse=True)
print('There are %d features having missing values'%len(miss_data))
print("Top two features in terms of missing values are '%s' and '%s'"%(miss_data[0][1],miss_data[1][1]))
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
```

```python
for i in data[cat_var].columns :
    data[i] = data[i].replace(-1,1000)

data[cat_var].describe()
```

```python
data_enc = pd.DataFrame(data[cat_var])
for i in data_enc.columns :
    t = enc.fit_transform(data[[i]]).toarray()
    data_enc.drop([i],axis=1,inplace=True)
    x = len(t[0])
    col_names = []
    for j in range(x) :
        col_names.append(i+'_'+str(j+1))
    data_enc = data_enc.join(pd.DataFrame(t,columns=col_names))

data_enc
```

```python
data[float_var].describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
t = scaler.fit_transform(data[float_var])
data_scale = pd.DataFrame(t,columns=float_var)
data_scale.describe()
```

```python
pd.DataFrame(data[bin_var],data_enc)
```

```python
y = data['target']
X = data.drop(['target'],axis=1)

X_new = pd.DataFrame(pd.DataFrame(data[[bin_var]],data_enc),data_scale)
y_new = y

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
X_new_train, X_new_test, y_new_train, y_new_test = train_test_split(X_new, y_new, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_fscore_support

clf = LogisticRegression(random_state=10).fit(X_train, y_train)
clf.score(X_test,y_test)

precision_recall_fscore_support(y_test,clf.predict(X_test))

clf_new = LogisticRegression(random_state=10).fit(X_new_train, y_new_train)
clf.score(X_new_test,y_new_test)

precision_recall_fscore_support(y_new_test,clf_new.predict(X_new_test))
```
