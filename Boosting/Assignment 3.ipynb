```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('letterCG.bin',sep=' ')
data = data.dropna(axis=1)
data
```

```python
X = data.drop(['Class'],axis=1)
y = data.Class
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
accuracy = []
for estimators in range(1,17) :
    model = AdaBoostClassifier(n_estimators=estimators).fit(X_train,y_train)
    accuracy.append(model.score(X_test,y_test))
plt.plot(accuracy)
```

```python
accuracy = []
for estimators in range(1,17) :
    model = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=2),n_estimators=estimators).fit(X_train,y_train)
    accuracy.append(model.score(X_test,y_test))
plt.plot(accuracy)
```
