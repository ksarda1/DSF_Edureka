```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
raw_data = pd.read_csv('run_or_walk.csv')
raw_data
```

```python
raw_data.columns
```

```python
raw_data.describe()
```

```python
raw_data.isna().sum()
```

```python
X = raw_data[['acceleration_x','acceleration_y','acceleration_z','gyro_x','gyro_y','gyro_z']]
y = raw_data['activity']
```

```python
clf = GaussianNB()
clf.fit(X, y)
clf.score(X,y)
```

```python
X_a = raw_data[['acceleration_x','acceleration_y','acceleration_z']]
clf.fit(X_a, y)
clf.score(X_a,y)
```

```python
X_g = raw_data[['gyro_x','gyro_y','gyro_z']]
clf.fit(X_g, y)
clf.score(X_g,y)
```
